```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dob_sol_times = np.loadtxt("dob_solution_times.txt")
mlp_sol_times = np.loadtxt("mlp_solution_times.txt")
```

```python
#TODO these must be extended with new results once experiments terminate. 

plt.figure(1, figsize=(12,6))

plt.subplot(121)
plt.title("Runtime for 2 houses")
plt.xlabel("Number of adversaries")
plt.ylabel("Run Time in Seconds")
dob = plt.semilogy(range(1,6), dob_sol_times[0,:],'v-', label='Dobbs')
mlp = plt.semilogy(range(1,6),mlp_sol_times[0,:],'*-', label='Multiple LPs')

# legend plot
plt.legend()

plt.subplot(122)
plt.title("Runtime for 3 houses")
plt.xlabel("Number of adversaries")
plt.ylabel("Run Time in Seconds")
dob = plt.semilogy(range(1,6), dob_sol_times[1,:],'v-', label='Dobbs')
mlp = plt.semilogy(range(1,6),mlp_sol_times[1,:],'*-', label='Multiple LPs')

# legend plot
plt.legend()

plt.show()
```
