```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dob_sol_times = np.loadtxt("dob_solution_times")
mlp_sol_times = np.loadtxt("mlp_solution_times")
```

```python
NUM_M, NUM_A = dob_sol_times.shape
print(NUM_M, NUM_A)
```

```python
# TODO: Logscale on the y axis. 

plt.figure(1, figsize=(12,12))

plt.subplot(221)
plt.title("Runtime for 2 houses")
plt.xlabel("Number of adversaries")
plt.ylabel("Run Time in Seconds")
dob = plt.plot(range(1,15), dob_sol_times[0,:] , label='Dobbs')
mlp = plt.plot(range(1,15),mlp_sol_times[0,:] ,label='Multiple LPs')

plt.legend([dob, mlp], ['Dobbs', 'Multiple LPs'])


plt.subplot(222)
plt.title("Runtime for 3 houses")
plt.xlabel("Number of adversaries")
plt.ylabel("Run Time in Seconds")
dob = plt.plot(range(1,15), dob_sol_times[1,:] , label='Dobbs')
mlp = plt.plot(range(1,15),mlp_sol_times[1,:] ,label='Multiple LPs')

plt.legend([dob, mlp], ['Dobbs', 'Multiple LPs'])


plt.subplot(223)
plt.title("Runtime for 4 houses")
plt.xlabel("Number of adversaries")
plt.ylabel("Run Time in Seconds")
dob = plt.plot(range(1,15), dob_sol_times[2,:] , label='Dobbs')
mlp = plt.plot(range(1,15),mlp_sol_times[2,:] ,label='Multiple LPs')

plt.legend([dob, mlp], ['Dobbs', 'Multiple LPs'])



plt.subplot(224)
plt.title("Runtime for 5 houses")
plt.xlabel("Number of adversaries")
plt.ylabel("Run Time in Seconds")
dob = plt.plot(range(1,15), dob_sol_times[3,:] , label='Dobbs')
mlp = plt.plot(range(1,15),mlp_sol_times[3,:] ,label='Multiple LPs')

plt.legend([dob, mlp], ['Dobbs', 'Multiple LPs'])

plt.show()
```
